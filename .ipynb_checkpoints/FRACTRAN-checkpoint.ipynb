```python
frac = np.array([[17,91],[78,85],[19,51],[23,38],[29,33],[77,29],[95,23],[77,19],[1,17],[11,13],[13,11],[15,2],[1,7],[55,1]])
```

```python
%time

magnum = 2
outputs = [2]
frac_list = [-1]
whilecount = 0
whilemax = 1500
while len(outputs) < whilemax:
    whilecount += 1
    arcount = 0
    for ar in frac:
        arcount += 1
        testnum = int(magnum) * int(ar[0]) / int(ar[1]) #calculate the product of that item (fraction) and the input
        print('testnum',testnum,'for ',str(ar[0])+'/'+str(ar[1]),'magnum',magnum)
        if testnum % 1 == 0:
            magnum = int(testnum)
            outputs.append(magnum)
            frac_list.append(arcount)
            print('-Found a match:',magnum,'...Breaking after arcount =',arcount, 'completed',whilecount,'of',whilemax)
            break
```

```python
i = 0
frac=list(frac)
acc = 0
for ff in frac_list:
    act_ratio = outputs[i+1] / outputs[i]
    des_ratio = frac[ff-1][0] / frac[ff-1][1]
    if abs((act_ratio / des_ratio)-1) > acc:
        print('discrepancy reported at',i,'/',
              outputs[i],'/',
              frac_list[i])
        break
    i += 1
```
