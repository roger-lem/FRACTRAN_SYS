```python
def moreprimes(count):
    global checkmax, primelist
    if count < checkmax:
        print('running moreprimes from',checkmax,'to',count)
    else:
        print('no need to run moreprimes')
    for num in range(max(primelist),count+1):
        is_prime = None
        for prime in primelist:
            if num % prime == 0:
                is_prime = False
        if is_prime == None:
            primelist.append(num)
        if checkmax < num:
            checkmax = num
```

```python
def pFactorize(entry):
    rem = entry
    factors = []
    if entry > checkmax:
        moreprimes(entry)
    for prime in primelist:
        while rem % prime == 0:
            rem /= prime
            factors.append(prime)
    return factors
```

```python
def FracFactorize(arr): #takes in an array of number pairs and returns 
    factarr = [[None,None] for i in range(len(arr))]
    arrindex = 0
    for pair in factarr:
        factarr[arrindex][0] = pFactorize(arr[arrindex][0])
        factarr[arrindex][1] = pFactorize(arr[arrindex][1])
        arrindex += 1
    return factarr
```

```python
def listProd(myList): 
    result = 1
    for x in myList: 
         result = result * x  
    return result
```
